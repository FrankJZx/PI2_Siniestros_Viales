#### Cambios en los datos de cada columna

| Columna              | Cambio                                                                                         |
|----------------------|------------------------------------------------------------------------------------------------|
| Id                   | Se eliminará el elemento "-" para convertir la columna en tipo entero (integer)               |
| Fecha                | Se eliminará el horario del formato, mostrando solo la fecha                                  |
| Hora                 | Se encontraron unos pocos valores atipicos como ej:"1900-01-02T00:00:00.000" por lo que se procedera a corregirlos    |
| Franja_Horaria       | Se cambiará a un formato:                                                                          |
|                      | - Mañana (6:01 a 12:00)                                                                         |
|                      | - Tarde (12:01 a 19:00)                                                                         |
|                      | - Noche (19:01 a 24:00)                                                                         |
|                      | - Madrugada (24:01 a 6:00)                                                                      |
| Calle                | Se encontró un valor nulo en la columna, por lo que se cambiará a "SD" (Sin Dato)            |
| Altura               | Se cambiarán los valores faltantes por "cruce_de_calles", ya que se encontró una correlación entre la columna "Altura" y "Cruce". Si existe dato en una columna, falta en la otra. |                         
| Cruce                | Se cambiarán los datos:                                                                          |
|                      | - Cuando sea un cruce, por "si"                                                                |
|                      | - Cuando sea faltante, por "no"                                                                |
| Direccion Normalizada| Se eliminará la columna, ya que hay otra columna que proporciona los mismos datos              |
| XY (CABA)            | Se eliminará la columna, ya que hay otras dos columnas que proporcionan los mismos datos     |



```python
############  Procedemos a realizar los cambios propuestos ############ 

# cambio columna ID
homicidios_hechos['Id'] = homicidios_hechos['Id'].str.replace('-', '')
homicidios_hechos['Id'] = homicidios_hechos['Id'].astype(int) 

# cambios columna Fecha
homicidios_hechos['Fecha'] = homicidios_hechos['Fecha'].dt.strftime('%Y-%m-%d')

# cambios en la columna Hora
def convertir_a_time(x):
    '''
    Convierte un valor a un objeto de tiempo (time) de Python si es posible.

    Esta función acepta diferentes tipos de entrada y trata de convertirlos en objetos de tiempo (time) de Python.
    Si la conversión no es posible, devuelve None.

    Parameters:
        x (str, datetime, or any): El valor que se desea convertir a un objeto de tiempo (time).

    Returns:
        datetime.time or None: Un objeto de tiempo (time) de Python si la conversión es exitosa,
        o None si no es posible realizar la conversión.
    '''
    if isinstance(x, str):
        try:
            return datetime.strptime(x, "%H:%M:%S").time()
        except ValueError:
            return None
    elif isinstance(x, datetime):
        return x.time()
    return x

# Se cambia el tipo de dato
homicidios_hechos['Hora'] = homicidios_hechos['Hora'].apply(lambda x: convertir_a_time(x))

# se observa en la fila 518 que quedo un valor faltante por lo que se asignara el valor 00:00:00 para no tener faltantes.
homicidios_hechos["Hora"].fillna("00:00:00", inplace=True)

#cambios en la columna Franja Horaria
def convertir_franja_horaria(valor):
  if isinstance(valor, int):
    if valor > 6 and valor <= 12:
      return "Mañana"
    elif valor > 12 and valor <= 19:
      return "Tarde"
    elif valor > 19 and valor <= 24:
      return "Noche"
    else:
      return "Madrugada"
  else:
    return valor

# Aplica la función a la columna "Franja_Horaria"
homicidios_hechos["Franja_Horaria"] = homicidios_hechos["Franja_Horaria"].apply(convertir_franja_horaria)

# cambios en la columna Calle
homicidios_hechos["Calle"].fillna("SD", inplace=True)

# cambios en la columna Altura
homicidios_hechos["Altura"].fillna("cruce_de_calles", inplace=True)

# cambios en la columna Cruce
def convertir_columna_Cruce(valor):
  if isinstance(valor, str):
    return "si"
  else:
    return "no"

# Aplica la función a la columna
homicidios_hechos ["Cruce"] = homicidios_hechos["Cruce"].apply(convertir_columna_Cruce)

# cambios en las columnas Dirección Normalizada y XY (CABA)
# al final decidi eliminar la columna cruce ya que los datos ya quedarian integrados en la columna altura
homicidios_hechos = homicidios_hechos.drop(columns=["Dirección Normalizada", "XY (CABA)", "Cruce"])

#################################################################################
""" 
LOS SIGUIENTES CAMBIOS NO FUERON AGREGADOS A LA LISTA DE MODIFICACIONES 
YA QUE ERAN PRINCIPALMENTE PARA LA VISUALIZACION DEL DASHBOARD EN POWER BI.
"""
#################################################################################

# agrego una columna para la visualizacion posterior en el dashboard

# Convertir la hora a string para asegurarse de que todos los valores sean de tipo string
homicidios_hechos['Hora'] = homicidios_hechos['Hora'].astype(str)

# Extraer la hora del día de la columna 'Hora'
homicidios_hechos['Hora del día'] = pd.to_datetime(homicidios_hechos['Hora']).dt.hour

# Contar la cantidad de homicidios por hora del día
conteo_por_hora = homicidios_hechos['Hora del día'].value_counts().sort_index()

###################################################################################

# Crear una nueva columna para almacenar la conversión de la columna 'Fecha' a datetime
homicidios_hechos['Fecha_Convertida'] = pd.to_datetime(homicidios_hechos['Fecha'])

# Obtener el nombre del día de la semana
homicidios_hechos['Dia_Semana'] = homicidios_hechos['Fecha_Convertida'].dt.day_name()

# Definir el tipo de día basado en el día de la semana
homicidios_hechos['Tipo_Dia'] = homicidios_hechos['Dia_Semana'].apply(lambda x: 'Dia_de_Semana' if x in ['Thursday', 'Friday', 'Monday', 'Tuesday', 'Wednesday'] else 'Fin_de_Semana')

# Reordenar el DataFrame según el orden de los días de la semana
dias_semana_ordenados = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
homicidios_hechos = homicidios_hechos.sort_values('Fecha_Convertida')
homicidios_hechos['Dia_Semana'] = pd.Categorical(homicidios_hechos['Dia_Semana'], categories=dias_semana_ordenados, ordered=True)

# se elimina la columna copiada
homicidios_hechos = homicidios_hechos.drop(columns=["Fecha_Convertida"])

# Mostrar el DataFrame
homicidios_hechos
```
